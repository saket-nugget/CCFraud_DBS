```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
```

```python
file_path = "creditcard.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
df.info()
```

```python
df = df.dropna()
df = df.drop(['Time','Class'],axis=1)
```

```python
df_sample = df.sample(frac=0.1, random_state=42).copy()
X = StandardScaler().fit_transform(df_sample)
```

```python
dbs = DBSCAN(eps=5.0,min_samples=5,metric='euclidean',n_jobs=-1)
```

```python
df_sample['cluster'] = dbs.fit_predict(X)
```

```python
df_sample['cluster'].value_counts()
```
